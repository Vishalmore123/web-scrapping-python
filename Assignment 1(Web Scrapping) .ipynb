```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
url = "https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles%7CMobiles&requestId=1495f0dc-9445-4225-b083-d5d5a4f9f436"
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html5lib')
```

```python
name = soup.find_all(class_='s1Q9rs')
for i in range (0,len(name)):
    name[i] = name[i].text
name
```

```python
prc = soup.find_all(class_='_30jeq3')
for i in range (0,len(prc)):
    prc[i] = prc[i].text[1:]
prc

rating = soup.find_all(class_= '_3LWZlK')
for i in range (0, len(rating)):
    rating[i] = rating[i].text
rating
```

```python
d = {"Book Name":name, "Price":prc[:40], "Ratings":rating[:40]}
df = pd.DataFrame(d)
df
```

```python
df.to_csv('Assignmen1.csv')
```
